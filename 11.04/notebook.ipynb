### Таск 1 и 2: открыть файл, отсортировать все по количеству треков(уникальных) и вывести топ 1

```python
data = []
art_songs = {}
with open('dataset.csv', 'r') as f:
    data = DictReader(f)
    for row in data:
        artists = row['artists'].split(';')
        for artist in artists:
            artist = artist.strip()
            if artist in art_songs:
                art_songs[artist].add(row['track_name'])
            else:
                art_songs[artist] = {row['track_name']}
    art_songs_sorted = sorted(art_songs.items(), key = lambda x : len(x[1]),reverse = True)
    for i in art_songs_sorted:
        print(i)
        break


    
```

### Таск 3: выбрал у исполнителей треки с популярностью больше 67 и отсортировал по количеству таких треков. максимальная популярность примерно 80, поэтому можно взять примерно такой промежуток для анализа
### для лузеров взял самый популярный трек каждого и отсортировал

```python
from csv import DictReader

data = []
art_songs = {}
art_hits = {}
art_max_pop = {}
with open('dataset.csv', 'r') as f:
    data = DictReader(f)
    for row in data:
        popularity = int(row['popularity'])
        artists = row['artists'].split(';')

        for artist in artists:
            artist = artist.strip()
            if artist in art_songs:
                art_songs[artist].add((row['track_name'], popularity))
            else:
                art_songs[artist] = {(row['track_name'], popularity)}
for artist in art_songs:
    hit_count = 0
    max_popularity = 0

    for song_name, popularity in art_songs[artist]:
        if popularity > 67:
            hit_count += 1
        if popularity > max_popularity:
            max_popularity = popularity

    art_hits[artist] = hit_count
    art_max_pop[artist] = max_popularity

art_hits = sorted(art_hits.items(), key=lambda x: x[1], reverse=True)
art_max_pop_sorted = sorted(art_max_pop.items(), key=lambda x: x[1])
print("hit makers\n")
for artist, hits in art_hits[:10]:
    print(artist, hits)
print("\nlooooooosers\n")
for artist, hits in art_max_pop_sorted[:10]:
    print(artist, hits)
```
